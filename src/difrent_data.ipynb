we need to fix and change oversample data

```python
test_df['quality'].value_counts(normalize=True)
```

```python
train_main_quality = train_df[train_df['quality']==5]
train_3_quality = train_df[train_df['quality']==3]
train_4_quality = train_df[train_df['quality']==4]
train_6_quality = train_df[train_df['quality']==6]
train_7_quality = train_df[train_df['quality']==7]
train_8_quality = train_df[train_df['quality']==8]




```

```python
total_other_rows = len(train_3_quality) + len(train_4_quality) + len(train_6_quality) + len(train_7_quality) + len(train_8_quality)

resampled__quality = train_main_quality.sample(n = round((total_other_rows) * 0.9),
                          replace=True)
```

```python
resampled__quality
```

```python
oversampled_train_df = pd.concat([train_main_quality,resampled__quality])

oversampled_train_df['quality'].value_counts(normalize=True)
```
