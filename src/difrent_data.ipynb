```python
x_train = oversampled_train_df.drop(columns=['quality'])
x_test = test_df.drop(columns=['quality'])

y_train = oversampled_train_df['quality']
y_test = test_df['quality']
```

```python
from sklearn.metrics import accuracy_score,recall_score,f1_score,classification_report,confusion_matrix,precision_score
```

```python
def model_evulator(x_matrix,y_matrix,model):
    preds=model.predict(x_matrix)
    print(f'accuracy_score : {accuracy_score(y_true=y_matrix, y_pred=preds)}')
    print(f'recall_score : {recall_score(y_true=y_matrix, y_pred=preds, average="weighted")}')
    print(f'f1_score : {f1_score(y_true=y_matrix, y_pred=preds, average="weighted")}')
    print(f'precision_score : {precision_score(y_true=y_matrix, y_pred=preds, average="weighted")}')
    print(f'confusion_matrix : \n{confusion_matrix(y_true=y_matrix, y_pred=preds)}')
    print(f'classification_report : \n{classification_report(y_true=y_matrix, y_pred=preds)}')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(x_train,y_train)
```

```python
model_evulator(x_train,y_train,rf_model)
```

```python
model_evulator(x_test,y_test,rf_model)
```

```python
from sklearn.model_selection import GridSearchCV 
```

```python
hyperparameters_dict ={


    "n_estimators":[50,100,175],
    "min_samples_split" : [2,4],
    "min_samples_leaf" : [4,8],
    "max_depth" :[2,4],
    "class_weight": ['balanced']
}
```

```python
Gridsearch_model = GridSearchCV(
    estimator = RandomForestClassifier(),
    param_grid = hyperparameters_dict,
    scoring = "accuracy",
    verbose = 3
    )
```

```python
Gridsearch_model.fit(x_train,y_train)
```

```python
model_evulator(x_train,y_train,Gridsearch_model)
```

```python
model_evulator(x_test,y_test,Gridsearch_model)
```
